```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

from bs4 import BeautifulSoup

Cities = ["San Diego, CA", "Salt Lake City, UT", "Houston, TX", "Boston, MA", "New York City, NY", "Zion National Park"]

def GetTemps(Cities):
    Website = Initiate("http://www.wunderground.com")
    
    try:
        for city in Cities:
        
            Search = Website.find_element_by_id("wuSearch")
            Search.send_keys(city, Keys.ENTER)
        
            Soup = BeautifulSoup(Website.page_source)
        
            Temp = Soup.find(id="curTemp")
            print city, Temp.span.span.text
        Website.close()
            
    except:
        print "Something happened"
        Website.close()
        
def Initiate(Site):
    Website = webdriver.Firefox()
    Website.get(Site)
    return Website
```

```python
def HistoricWeather(City, date):
    Website = Initiate("http://www.wunderground.com/history/")
    
    month = date[0]
    print month
    day = date[1]
    print day
    
#    try:
    Search = Website.find_element_by_id("histSearch")
    Search.clear()
    Search.send_keys(City)
    
    tot_years = len(Website.find_element_by_class_name('year').find_elements_by_tag_name('option'))
    
    for year in range(tot_years):
        
        Month = Website.find_element_by_class_name("month")
        ExactMonth = Month.find_elements_by_tag_name("option")[month-1]
        ExactMonth.click()
        
        Day = Website.find_element_by_class_name("day")
        ExactDay = Day.find_elements_by_tag_name("option")[day-1]
        ExactDay.click()
        
        Year = Website.find_element_by_class_name("year")
        ExactYear = Year.find_elements_by_tag_name("option")[year]
        ExactYear.click()
        
        try:
            Submit = Website.find_element_by_class_name("location-date-column").find_element_by_tag_name("input")
        except:
            Submit = Website.find_element_by_class_name("history-date-select").find_element_by_tag_name("input")
        
        Submit.submit()
        
        Soup = BeautifulSoup(Website.page_source)
        
        print (2015-year), Soup.find(text="Max Temperature").parent.parent.next_sibling.next_sibling.span.span.text
        
    Website.quit()
```
