```python
import matplotlib.pyplot as plt
N = 1000
s = np.zeros(N)
s[1:] = np.random.binomial(1, .5, size=(N-1,))*2-1
s = pd.Series(s)
s = s.cumsum()
s.plot()
plt.ylim([-50,50])
plt.show()
plt.close()
```

```python
#problem 2
N = 100
for x in range(5):
    s = np.zeros(N)
    s[1:] = np.random.binomial(1, .5, size=(N-1,))*2-1
    s = pd.Series(s)
    s = s.cumsum()
    s.plot()
plt.ylim([-50,50])
plt.show()
plt.close()
```

```python
#build toy data for SQL operations
name = ['Bill','Alice','Joe','Jenny','Ted','Taylor','Tracy','MMorgan','Liz']
sex = ['M','F','M','F','M','F','M','M','F']
age = [20, 21, 18, 22, 19, 20, 20, 19, 20]
rank = ['Sp','Se','Fr','Se','Sp','J','J','J','Se']
ID = range(9)
aid = ['y','n','n','y','n','n','n','y','n']
GPA = [3.8,3.5,3.0,3.9,2.8,2.9,3.8,3.4,3.7]
mathID = [0,1,5,6,3]
mathGd = [4.0,3.0,3.5,3.0,4.0]
major = ['y','n','y','n','n']
studentInfo = pd.DataFrame({'ID':ID,'Name':name,'Sex':sex,'Age':age,'Class':rank})
otherInfo = pd.DataFrame({'ID':ID,'GPA':GPA,'Financial_Aid':aid})
mathInfo = pd.DataFrame({'ID':mathID,'Grade':mathGd,'Math_Major':major})
```

```python
import pandas as pd
Crime = pd.read_csv('crime_data.txt', header=1, index_col='Year')
Crime
```

```python
import matplotlib.pyplot as plt
plt.plot(Crime.index,Crime['Crime Rate'])
plt.show()
plt.close()
Crime.plot(y='Crime Rate')
plt.show()
plt.close()
```
